```python
x = np.linspace(0, 2*np.pi)
fig, ax = plt.subplots(3, 3)
ax1, ax2, ax3 = ax
ax11, ax12, ax13 = ax1
ax21, ax22, ax23 = ax2
ax31, ax32, ax33 = ax3

fig.set_figwidth(8)
fig.set_figheight(8)

ax11.plot(x, np.sin(x))
ax12.plot(x, np.cos(x))
ax13.plot(x, np.tan(x))

ax21.plot(x, np.sin(x))
ax22.plot(x, np.cos(x))
ax23.plot(x, np.tan(x))

ax31.plot(x, np.sin(x) + np.cos(x))
ax32.plot(x, np.sin(x))
ax33.plot(x, np.sin(x))

plt.tight_layout()
```

```python
plt.figure(figsize=(6, 4))
x = np.linspace(0, 10, 100)

axes1 = plt.gca()
axes1.plot(x, np.exp(x), c="r")
axes1.set_xlabel("time")
axes1.set_ylabel("exp", c="r")
axes1.tick_params(axis="y", labelcolor="red")

axes2 = axes1.twinx()
axes2.plot(x, np.sin(x), c="b")
axes2.set_ylabel("sin", c="b")
axes2.tick_params(axis="y", labelcolor="blue")

plt.tight_layout()
```

```python
plt.figure(figsize=(6, 4))

x = np.linspace(0, 2*np.pi)
plt.plot(x, np.sin(x), label="sin")
plt.plot(x, np.cos(x), label="cos")
plt.legend()
```

```python
plt.figure(figsize=(6, 4))

x = np.linspace(0, 2*np.pi)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.legend(["sin", "cos"], fontsize=18, ncol=2, bbox_to_anchor=[0, 0, 1, 0.2])  # loc="best"
```

```python
plt.figure(figsize=(6, 4))

x = np.linspace(0, 2*np.pi, 20)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, color="r", marker="o", ls="--", label="sin(x)", mfc="y", lw=1)  # ls: line style; lw: line width; mfc: marker facecolor
plt.plot(x, y2, color="b", marker="*", ls="-", label="cos(x)", mfc="w", lw=2, markersize=10) 
plt.plot(x, y1-y2, color="y", marker="^", ls="-", label="sin(x)-cos(x)", mfc="b", lw=2, markersize=10, alpha=0.3) 
plt.legend(["sin", "cos", "sin-cos"], fontsize=18)  # loc="best"
```
